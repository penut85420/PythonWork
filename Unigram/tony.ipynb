```python
f = open('lexicon_test.txt', encoding = 'utf8')
lexicon = list()
for line in f.readlines():
    lexicon.append(str.split(line))

print(len(lexicon))
print(lexicon)
```

```python
f = open('wsegInput_test.txt', encoding = 'utf8')
wsegInput = list()
for line in f.readlines():
    wsegInput.append(str.split(line))
    
print(len(wsegInput))
print(wsegInput)
```

```python
new_wseg = []
for sentence in wsegInput:
    print(sentence)
    i = 0
    this_sen = []
    while i < len(sentence[0]):
        # 如果lexicon有這個字
        if [sentence[0][i]] in lexicon: 
            #print(sentence[0][i])
            this_sen.append([sentence[0][i]])
            print("", end = "")
        
        # lexicon沒這個字 => 
        # 接一個字之後再去lexicon找
        else:
            print(sentence[0][i])
            word = ""
            print("i =", i)
            while ([word] not in lexicon):
                print(word + " not in lexicon")
                word = word + sentence[0][i]
                i = i + 1
            i = i - 1
            this_sen.append([word])
            print(word + " in lexicon !!!")
            print("i =", i)
        
        i = i + 1
    print("this_sen =",this_sen)
    new_wseg.append(this_sen)
```

```python
def find_word_2(i, start, pre_str):
    # 找單字詞
    my_str = new_wseg[i][start][0]
    if [my_str] in lexicon:
        if start+1 < len(new_wseg[i]):
            print(my_str, end = " ")
            find_word_2(i, start+1, pre_str + my_str + " ")
        else:
            print(my_str, end = "\n")
            
    
    # 找二字詞
    if start+1 < len(new_wseg[i]):
        my_str = new_wseg[i][start][0] + new_wseg[i][start+1][0]
        if [my_str] in lexicon:
            if start+2 < len(new_wseg[i]):
                print(pre_str+my_str, end = " ")
                find_word_2(i, start+2, pre_str + my_str + " ")
            else:
                print(pre_str+my_str, end = "\n")
                
    
    # 找三字詞
    if start+2 < len(new_wseg[i]):
        my_str = new_wseg[i][start][0] + new_wseg[i][start+1][0]+ new_wseg[i][start+2][0]
        if [my_str] in lexicon:
            if start+3 < len(new_wseg[i]):
                print(pre_str+my_str, end = " ")
                find_word_2(i, start+3, pre_str + my_str + " ")
            else:
                print(pre_str+my_str, end = "\n")
```
